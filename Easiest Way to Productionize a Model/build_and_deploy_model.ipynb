```python
import pandas as pd
from pycaret.classification import *
```

```python
df = df.drop(["Loan ID", "Customer ID"], axis = 1)
```

```python
df.columns = ['Loan_Status', 'Current_Loan_Amount', 'Term', 'Credit_Score',
       'Annual_Income', 'Years_in_current_job', 'Home_Ownership', 'Purpose',
       'Monthly_Debt', 'Years_of_Credit History',
       'Months_since_last_delinquent', 'Number_of_Open_Accounts',
       'Number_of_Credit_Problems', 'Current_Credit_Balance',
       'Maximum_Open_Credit', 'Bankruptcies', 'Tax_Liens']
```

```python
df["Loan_Status"].value_counts()
```

```python
df["Term"].value_counts()
```

```python
def change_name(s):
    if s == "< 1 year":
        return "1 year"
    else:
        return s
    
df["Years_in_current_job"] = df["Years_in_current_job"].map(change_name)
```
